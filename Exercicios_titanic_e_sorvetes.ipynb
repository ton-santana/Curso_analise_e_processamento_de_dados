## Exercício 01 - Analisando Dataset do Titanic.

```python
# importando biblioteca pandas
import pandas as pd
```

```python
# importando dataset com informações do Titanic.
data_titanic = pd.read_csv("datasets/Datasets_e_Dataframes-20210824/titanic.csv", sep = ",")
```

```python
# mostrando os dados do dataFrame
# somente as 5 primeiras linhas
data_titanic.head()
```

```python
# escolhendo colunas a serem visualizadas.
dados = data_titanic[["Name", "Age", "Sex", "Pclass", "Survived"]]
# mostrando resultados
dados
```

```python
# Mostrando passageiros do sexo feminino da primeira classe que sobreviveram
passageirosF = dados.loc[(dados["Sex"] == "female") & (dados["Pclass"] == 1) & (dados["Survived"] == 1)]
# exibindo o resultado do filtro.
passageirosF
```

```python
# Filtrando a idade dos passageiros do sexo masculino sobreviventes da 3ª classe em ordem crescente.
passageiros_3classe = dados.loc[(dados["Sex"] == "male") & (dados["Pclass"] == 3) & (dados["Survived"] == 1)].sort_values(["Age"], ascending = True)
passageiros_3classe
```

```python
# visualizando todos os dados em ordem alfabética.
dados_ordenados = dados.sort_values(["Name"], ascending = True)
dados_ordenados
```

```python
# Exibindo histograma de idades dos passageiros do Titanic.
dados_ordenados.hist("Age", figsize = (10, 8))
```

* Através do histograma podemos ver que a maioria dos passageiros possui entre 18 e 32 anos, aproximadamente.

```python
# Analisar a quantidade de sobreviventes por sexo.
qtd_sobreviventes_por_sexo = dados_ordenados[['Sex', 'Survived']].groupby(['Sex']).sum()
qtd_sobreviventes_por_sexo
```

```python
# Exibindo a quantidade de sobreviventes por sexo em um gráfico de barras.
qtd_sobreviventes_por_sexo.plot.bar()
```

```python
# Exportando os dados no formato CSV.
dados_ordenados.to_csv("datasets/Datasets_e_Dataframes-20210824/dados_titanic_ordenados.csv", index = False)
```

## Exercício 02 - Analisando dados do Dataset de Sorvetes.

```python
# importando bibliotéca matplotlib.
import matplotlib.pyplot as plt
```

```python
# criando o dataFrame de sorvetes.
sorvetes = pd.read_csv("datasets/Datasets_e_Dataframes-20210824/sorvetes.csv", sep = ",")
# exibindo o dataFrame
sorvetes
```

* A coluna Temperatura está em graus Fahrenheit.

```python
# Criando uma nova coluna no dataFrame para adicionar a temperatura em graus Celcius.
sorvetes["Temperatura_celcius"] = round(((sorvetes["Temperature"] - 32) * 5) / 9)
# Exibindo o dataFrame
sorvetes
```

```python
# Ordenando os valores pela temperatura em Fahrenheit em ordem decrescente
sorvetes = sorvetes.sort_values(["Temperature"], ascending = False)
# exibindo os resultados
sorvetes
```

* Visualizando o DataFrame em um gráfico.

```python
# utilizando a biblioteca matplotlib
plt.plot(sorvetes["Temperature"], sorvetes["Sales"], "r.--", label = "US$")
plt.title("Analisando vendas")
plt.ylabel("Vendas em US$")
plt.xlabel("Temperatura em °F")
plt.legend()

# exportando o gráfico em formato pdf
plt.savefig("datasets/Datasets_e_Dataframes-20210824/grafico_sorvetes.pdf")

# exibindo o gráfico
plt.show()
```

* Exibindo o gráfico em graus Celcius.

```python
# utilizando a biblioteca matplotlib
plt.plot(sorvetes["Temperatura_celcius"], sorvetes["Sales"], "g.--", label = "US$")
plt.title("Analisando vendas")
plt.ylabel("Vendas em US$")
plt.xlabel("Temperatura em °C")
plt.legend()

# exportando o gráfico em formato pdf
plt.savefig("datasets/Datasets_e_Dataframes-20210824/grafico_sorvetes_temperatura_celcius.pdf")

# exibindo o gráfico
plt.show()
```

```python
sorvetes.plot.scatter(x = "Temperatura_celcius", y = "Sales")
```

* Através do gráfico podemos verificar que quanto maior a temperatura em graus Celcius, maiores são as vendas.
